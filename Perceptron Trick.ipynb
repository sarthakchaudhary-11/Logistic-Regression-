# ðŸ“Œ Step-by-Step Perceptron Implementation

# Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Step 2: Generating Dataset

```python
# Generate a simple dataset
np.random.seed(42)
X = np.random.randn(100, 2)  # 100 samples, 2 features
y = np.where(X[:, 0] + X[:, 1] > 0, 1, -1)  # Linearly separable condition
```

# Step 3: Perceptron Learning Algorithm

```python

# Initialize weights and bias
w = np.random.randn(2)  # 2 weights
b = np.random.randn()   # Bias term
learning_rate = 0.1
epochs = 20  # Number of iterations

# Perceptron Trick Implementation
for epoch in range(epochs):
    for i in range(len(X)):
        y_pred = np.sign(np.dot(w, X[i]) + b)  # Prediction
        if y_pred != y[i]:  # If misclassified, update weights
            w += learning_rate * y[i] * X[i]  # Update weight
            b += learning_rate * y[i]        # Update bias
```

# Step 4: Plot Decision Boundary



```python


# Plot decision boundary
xx, yy = np.meshgrid(np.linspace(-3, 3, 100), np.linspace(-3, 3, 100))
Z = np.sign(w[0] * xx + w[1] * yy + b)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap="bwr", edgecolors='k')
plt.contour(xx, yy, Z, levels=[0], colors="black")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Perceptron Decision Boundary")
plt.show()

```
